```python
import os
import json

from src.solver.ceoh.methods.eoh_pooling.database import PopulationDatabase
from src.solver.ceoh.methods.eoh_pooling.pooling_class import CeohPooling
from src.solver.ceoh.methods.eoh_pooling.util import get_destroy_counts, get_destroy_scores, get_results_fitness

import random
base_path = "/home/nico/KIT_IMI/multibay-reshuffling/"

results_folder = os.path.join(base_path, "debug", "pooling_data", "sample_ceoh_data")
experiment_result = os.path.join(base_path, "debug", "pooling_data", "results.json")

db = PopulationDatabase(results_folder)

pooling = CeohPooling(db)
pooling.create_pools(0)


for p in pooling.pools:
    print(p.get_ids())


with open(experiment_result) as f:
    fitness, results = json.load(f)

for r in results:
    print(get_destroy_scores(r))

for i, p in enumerate(pooling.pools):
    p.set_pool_score(fitness * random.random())
    for p_i in range(8):
        p.set_program_score(p_i, fitness * random.random())

pooling.evaluate_and_update_pools(results)

print("##########################################")
print("Before evaluation:")

for p in pooling.pools:
    print(p.get_score())
    print(p.keep_pool())


print("##########################################")
print("After evaluation:")

for p in pooling.pools:
    print(p.get_score())
    print(p.get_program_scores())

pooling.mutate_pool(0)

print("##########################################")
print("After Mutate:")

for p in pooling.pools:
    print(p.get_score())
    print(p.get_program_scores())

print("##########################################")
print("After Another Eval:")

pooling.evaluate_and_update_pools(results)

for p in pooling.pools:
    print(p.get_score())
    print(p.get_program_scores())
```
